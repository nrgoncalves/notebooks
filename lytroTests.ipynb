# Dealing with Lytro images

Each Lytro examples gives us:
- a focus stack
- a depth map

In the end, we want to have 30x30 image patches that have either crossed or uncrossed disparity. 

These are the steps we need to follow for each acquisition:

    For each image in the focus stack:
        downsample?
        convert image to grayscale
        generate a stereo pair, and classify each patch as crossed vs uncrossed
        crop image in 30x30 patches
        for each path
            discard if
                zero disparity patches
                patches with low image variance (no features in image patch)
            store
                otherwise

The output of this process shall then be a set of 30x30 image patches labeled as crossed or uncrossed disparity. 

**Note:** Remember that we should generate stereopairs with small disparities (<< 30 pixels)!

## Generating stereo pairs

Each image-stack is associated with a depth map. The depth map, however, provides relative, not absolute, depth information. Therefore, we can start by defining the absolute disparity range. We will start by defining the maximum disparity magnitude to be 10 pixels (crossed and uncrossed).

As a starting point, we will generate two stereo pairs from each image. One stereopair will have crossed disparity (between -10 ans 0 pixels), while the other image will have uncrossed disparity (between 0 and 10 pixels).

The general idea would be

    for each image
        define max/min disparity (range)
        make position shift images from depth map
        find equivalend zero disparity in position shift images
        load corresponding image from focus stack
        generate stereo pair
        
However, we will only have all crossed and uncrossed images, so the procedure above will translate to:

    for each image
        for each disparity sign (crossed/uncrossed)
            if crossed
                dspRange = [-10,0]
            if uncrossed
                dspRange = [0,10]
            
            make position shift image by rescaling depth map to dspRange 
            (this is oversimplistic but ok for now - later we will do this properly by raytracing)
            
            if crossed
                stackId = last
            if crossed
                stackId = first

            load image[stackId]
            generate stereo pair

Ok, I think this makes sense. Let's demonstrate it now.

```python
import os, os.path

baseDir = "/Users/NunoGoncalves/Desktop/stereo-datasets/lfsd/"
depthDir = "/Users/NunoGoncalves/Desktop/stereo-datasets/lfsd/depth_map/"
focusDir = "/Users/NunoGoncalves/Desktop/stereo-datasets/lfsd/focus_stack/"

imageId = 1;
dspRange = np.array([[-10,0],[0,10]]) # first row for crossed, second for uncrossed

for ii in range(imageId):
    # get the depth map
    depthMap = mpimg.imread(depthDir + str(imageId) + '.bmp')
    
    # get however many focus images in this example
    nImages = len([name for name in os.listdir(imgDir) if (name.split("__")[0]==str(1))])
    
    for dd in range(np.shape(dspRange)[0]):
        posShift = np.double(depthMap)/np.amax(depthMap)*np.abs(dspRange[dd,1]-dspRange[dd,0])-np.amin(dspRange[dd,:])
        
        if dd == 0: # crossed, so load last image
            focusId = nImages-1
        elif dd == 1:
            focusId = 0
        
        img = mpimg.imread(focusDir + str(imageId) + '__refocus_' + ("%02d" % focusId) + '.jpg')
        
        gsImg = rgb2gray(img)
        fImg = imresize(gsImg, np.shape(depthMap), interp='bicubic')
        
        plt.subplot(1,2,dd+1)
        plt.imshow(fImg,cmap="gray")
        plt.title("Focus image " + str(focusId))

plt.show()
```

So, we are loading two images per plenoptic acquisition, each of which will give us one stereo-pair. In this case, image 9 will be rendered with crossed disparity (since we are focusing in the farmost point of the image), whereas image 0 will be rendered with uncrossed disparity. Let's now do the stereoscopic rendering (*note to self: must not forget to introduce ray-tracing to compute position shifts)*.


```python
# alright, let's have a go at the rendering
# we assume that we already computed position shifts
from scipy import ndimage

imgShape = np.shape(fImg)
rows = imgShape[0]
cols = imgShape[1]

limg = np.zeros((rows,cols))
lmsk = np.zeros((rows,cols))
rimg = np.zeros((rows,cols))
rmsk = np.zeros((rows,cols))

posShift = np.round(posShift/2)
maxShift = np.round(np.amax(posShift))
# maxShift = 5 # pixels


##################
### left image

# for every row
for rr in range(rows):
    
    # for every column, from end to start
    for cc in reversed(range(cols)):
        shift = posShift[rr, cc]

        if (cc-shift+maxShift < cols):
            limg[rr,cc-shift+maxShift] = fImg[rr,cc]
            lmsk[rr,cc-shift+maxShift] = 1

# this seems to work as a closing operation (dilation followed by erosion)
y = ndimage.grey_dilation(lmsk, size=(1,6), structure=np.array([[1,1,1,0,0,0]]))
x = ndimage.grey_erosion(y, size=(1,9), structure=np.array([[0,0,0,1,1,1,1,1,1]]))
lmsk *= x


for rr in range(rows):
    # filling disocclusions with Background Pixel Extrapolation
    for cc in range(cols):
        if (lmsk[rr,cc] == 0):
            if (cc-7<0):
                limg[rr,cc] = fImg[rr,cc]
            else:
                limg[rr,cc] = np.mean(limg[rr,cc-7:cc-3])

for rr in range(3,rows-2):
    # smooth filled occlusions
    for cc in range(3,cols-2):
        if (lmsk[rr, cc]==0):
            limg[rr, cc] = np.mean(limg[rr-3:rr+2,cc-3:cc+2])


##################
### right image

# for every row
for rr in range(rows):
    
    # for every column, from end to start
    for cc in reversed(range(cols)):
        shift = posShift[rr, cc]

        if (cc+shift-maxShift > 1):
            rimg[rr,cc+shift-maxShift] = fImg[rr,cc]
            rmsk[rr,cc+shift-maxShift] = 1

# this seems to work as a closing operation (dilation followed by erosion)
y = ndimage.grey_dilation(rmsk, size=(1,6), structure=np.array([[1,1,1,0,0,0]]))
x = ndimage.grey_erosion(y, size=(1,9), structure=np.array([[0,0,0,1,1,1,1,1,1]]))
rmsk *= x

         
for rr in range(rows):
    # filling disocclusions with Background Pixel Extrapolation
    for cc in reversed(range(cols)):
        if (rmsk[rr,cc] == 0):
            if (cc+7>cols-1):
                rimg[rr,cc] = fImg[rr,cc]
            else:
                rimg[rr,cc] = np.mean(rimg[rr,cc+4:cc+8])


for rr in range(3,rows-2):
    # smooth filled occlusions
    for cc in reversed(range(3,cols-2)):
        if (rmsk[rr, cc]==0):
            rimg[rr, cc] = np.mean(rimg[rr-3:rr+2,cc-3:cc+2]);
         
```

```python
plt.subplot(1,2,1)
plt.imshow(limg, cmap="gray",interpolation="nearest")
plt.title("left image")

plt.subplot(1,2,2)
plt.imshow(rimg, cmap="gray",interpolation="nearest")
plt.title("right image")
plt.show()
```
